```python
from velplot import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_sensor(id):
    l = d.data[id]['Sample Name']+" ID="+d.data[id]['ID']
    plt.plot(d.data[id]['t']/60.0, d.data[id]['p'], 'o-', label=l)

def plot_sensors(d):
    for id in d.ids():
        l = d.data[id]['Sample Name']+" ID="+d.data[id]['ID']
        plt.plot(d.data[id]['t']/60.0, d.data[id]['p'], 'o-', label=l)

def plot_sensors_norm(d):
    for id in d.ids():
        factor = d.data[id]['p'][1] # ci≈õnienie startowe do normalizacji (pierwszy pomiar odrzucony)
        l = d.data[id]['Sample Name']+" ID="+d.data[id]['ID']
        plt.plot(d.data[id]['t'][1:]/60.0, d.data[id]['p'][1:]/factor, 'o-', label=l)

```

```python
d=Dataset("report.txt")
plot_sensors(d)
plt.xlim(0,2*24)
plt.xlabel("Czas, h")
plt.ylabel("Ci≈õnienie, mbar")
plt.legend()
```

```python

plot_sensors_norm(d)
plt.xlim(0,2*24)
plt.xlabel("Czas, h")
plt.ylabel("Ci≈õnienie, mbar")
plt.legend()
```
